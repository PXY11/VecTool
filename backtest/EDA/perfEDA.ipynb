```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def perf(path,name):
    df = pd.read_csv(f'{path}{name}.csv',index_col=0)
    df.columns =  ['tp'+str(col) for col in df.columns.tolist()]
    df.index = ['er'+str(ind) for ind in df.index.tolist()]
    df
    print(f'【{name}】')
    plt.figure(figsize=(10,6))
    sns.heatmap(df,annot=True ,linewidths=0.5,cmap="YlGnBu")
    plt.show()
```

# V6版本，5个长历史币种，每次选er最大1个币，增加er为正的进场条件

```python
path = '../testBBELB/perf/perf_v6/'
perf(path,'annualizedReturn')
perf(path,'sharpeRatio')
# perf(path,'maxDrawdown')
```
