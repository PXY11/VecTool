```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def perf(path,name):
    df = pd.read_csv(f'{path}{name}.csv',index_col=0)
    df.columns =  ['ar_param'+str(col) for col in df.columns.tolist()]
    df.index = ['winsize'+str(ind) for ind in df.index.tolist()]
    df
    print(f'【{name}】')
    plt.figure(figsize=(10,6))
    sns.heatmap(df,annot=True ,linewidths=0.5,cmap="YlGnBu")
    plt.show()
```

# V1版本，arRollNorm大于10时空仓

```python
path = '../testBBELB/perf/perf_ar_v1/'
perf(path,'annualizedReturn')
perf(path,'sharpeRatio')
# perf(path,'maxDrawdown')
```
